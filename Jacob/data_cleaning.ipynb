```python
import pandas as pd
```

```python
#read in files
```

```python
heart_df=pd.read_csv("resources/heart.csv")
heart_df.head()
```

```python
#check data types
```

```python
heart_df.dtypes
```

```python
#drop unecessary columns
```

```python
heart_df=heart_df.drop(columns=['Unnamed: 0'])
heart_df.head()
```

```python
#convert age to years 
```

```python
heart_df['age (years)'] = heart_df['age'].div(365)
heart_df.head()
```

```python
# change age (years) dtpye to integer
```

```python
heart_df['age (years)']=heart_df['age (years)'].astype('int')
heart_df.head()
```

```python
#check for null values
```

```python
heart_df.isnull().sum()
```

```python
# drop age column 
```

```python
heart_df=heart_df.drop(columns=['age'])
heart_df.head()
```

```python
cols = list(heart_df.columns.values)
cols
```

```python
heart_df=heart_df[['id',
 'age (years)','gender',
 'height',
 'weight',
 'ap_hi',
 'ap_lo',
 'cholesterol',
 'gluc',
 'smoke',
 'alco',
 'active',
 'cardio',]]
heart_df.head()
```

```python
#convert height and weight data from metric to U.S. units 
```

```python
heart_df['height (feet)'] = heart_df['height'].div(30.48)
heart_df.head()
```

```python
heart_df['height (feet)']=heart_df['height (feet)'].round(decimals = 2)
heart_df.head()
```

```python
heart_df['weight (pounds)'] = heart_df['weight']*2.205
heart_df.head()
```

```python
heart_df['weight (pounds)'] = heart_df['weight (pounds)'].round(decimals = 0)
heart_df.head()
```

```python
heart_df['weight (pounds)'] = heart_df['weight (pounds)'].astype('int')
```
