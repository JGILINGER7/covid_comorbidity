```python
#Import dependencies
```

```python
import pandas as pd
```

```python
#read in files
```

```python
heart_df=pd.read_csv("resources/heart.csv")
heart_df.head()
```

```python
#check data types
```

```python
heart_df.dtypes
```

```python
#drop unecessary columns
```

```python
heart_df=heart_df.drop(columns=['Unnamed: 0'])
heart_df.head()
```

```python
#convert age to years 
```

```python
heart_df['age (years)'] = heart_df['age'].div(365)
heart_df.head()
```

```python
# change age (years) dtpye to integer
```

```python
heart_df['age (years)']=heart_df['age (years)'].astype('int')
heart_df.head()
```

```python
#check for null values
```

```python
heart_df.isnull().sum()
```

```python
# drop age column 
```

```python
heart_df=heart_df.drop(columns=['age'])
heart_df.head()
```

```python
cols = list(heart_df.columns.values)
cols
```

```python
heart_df=heart_df[['id',
 'age (years)','gender',
 'height',
 'weight',
 'ap_hi',
 'ap_lo',
 'cholesterol',
 'gluc',
 'smoke',
 'alco',
 'active',
 'cardio',]]
heart_df.head()
```

```python
#convert height and weight data from metric to U.S. units 
```

```python
heart_df['height (feet)'] = heart_df['height'].div(30.48)
heart_df.head()
```

```python
heart_df['height (feet)']=heart_df['height (feet)'].round(decimals = 2)
heart_df.head()
```

```python
heart_df['weight (pounds)'] = heart_df['weight']*2.205
heart_df.head()
```

```python
heart_df['weight (pounds)'] = heart_df['weight (pounds)'].round(decimals = 0)
heart_df.head()
```

```python
heart_df['weight (pounds)'] = heart_df['weight (pounds)'].astype('int')
```

```python
heart_df.to_csv("heart_cleaned.csv", index= False)
```

```python
heart_df['inches']=heart_df['height'].div(2.54)
heart_df.head()
```

```python
heart_df['BMI'] = heart_df['weight (pounds)'].div(heart_df['inches']*heart_df['inches'])*703
heart_df.head()
```

```python
cols = list(heart_df.columns.values)
cols
```

```python
bmi_df=heart_df[[
 'id','age (years)','gender','BMI',
 'ap_hi',
 'ap_lo',
 'cholesterol',
 'gluc',
 'smoke',
 'alco',
 'active',
 'cardio']]
bmi_df.head()
```

```python
bmi_df.to_csv("bmi_simple.csv", index= False)
```

```python
heart_df.groupby('gender')[["weight (pounds)", "height (feet)"]].describe()
```

```python
#clean outliers from data drop ap outliers
```

```python
heart_df_cleaned = heart_df.loc[(heart_df["ap_hi"]>=90) & (heart_df["ap_lo"]>=60) & (heart_df["ap_hi"]<200) & (heart_df["ap_lo"]< 130) ]
heart_df_cleaned.head()
```

```python
# drop male height and weight outliers
```

```python
heart_cleaner_df = heart_df_cleaned [(heart_df_cleaned['gender']==2) & (heart_df_cleaned['weight (pounds)']>=95) & (heart_df_cleaned['weight (pounds)'] <=242.5)& (heart_df_cleaned['height (feet)']>=4.9) & (heart_df_cleaned['height (feet)']<=6.235)
|(heart_df_cleaned['gender']==1) & (heart_df_cleaned['weight (pounds)']>=83.5) & (heart_df_cleaned['weight (pounds)'] <=231.5)& (heart_df_cleaned['height (feet)']>=4.7) & (heart_df_cleaned['height (feet)']<=5.9)]

heart_cleaner_df.head(25)
```

```python
heart_cleaner_df.to_csv("resources/heart_cleaner.csv", index= False)
```
